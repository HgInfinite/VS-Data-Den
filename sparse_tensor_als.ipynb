```python
import numpy as np
import tensorly as tl
from tensorly.decomposition import parafac, tucker
```

```python
sales_tensor = np.load("sales_3d_array.npy")
```

```python
rank = 2
sparse_cp_factors = parafac(sales_tensor, rank=rank,n_iter_max=25,init='random', tol=1e-2, verbose=2, return_errors=True)
```
