```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
unique_articles = merged_df['ARTICLE'].unique()

```

```python
unique_articles
```

```python

article_arrays = {}

for article in unique_articles:

    article_df = merged_df[merged_df['ARTICLE'] == article]
    

    unique_regions = article_df['PTNR_REGIONAL_MASTER_NAME'].unique()
    unique_yearweeks = article_df['YEARWEEK'].unique()

    sales_array = np.zeros((len(unique_regions), len(unique_yearweeks)))
    

    for i, region in enumerate(unique_regions):
        for j, yearweek in enumerate(unique_yearweeks):
            sales = article_df[(article_df['PTNR_REGIONAL_MASTER_NAME'] == region) & (article_df['YEARWEEK'] == yearweek)]['SALES'].sum()
            sales_array[i, j] = sales
    
    article_arrays[article] = (unique_regions, unique_yearweeks, sales_array)
```

```python
for article, (regions, yearweeks, sales_array) in article_arrays.items():
    plt.figure(figsize=(10, 6))
    plt.imshow(sales_array, cmap='viridis', interpolation='nearest')
    plt.title(f'Sales Matrix for ARTICLE: {article}')
    plt.xlabel('YEARWEEK')
    plt.ylabel('PTNR_REGIONAL_MASTER_NAME')
    plt.xticks(range(len(yearweeks)), yearweeks, rotation=90)
    plt.yticks(range(len(regions)), regions)
    plt.colorbar(label='Sales')
    plt.tight_layout()
    plt.show()
```
