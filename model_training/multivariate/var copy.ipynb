```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np

from statsmodels.tsa.stattools import grangercausalitytests
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
```

```python
df = pd.read_parquet('/Users/harshitgupta/Desktop/vs /VS-Data-Den/model_training/multivariate/sorted_data.parquet')
```

```python
df.columns
```

```python
df = df.set_index('DATE_FIELD')
```

```python
df.head()
```

```python
df['FLAG'] = df['FLAG'].map({'Inactive': 0, 'Active': 1})
```

```python
max_lags = 12 
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
for code in df['PTNR_RM_CODE'].unique():
    print('PTNR_RM_CODE:', code)
    print('-' * 40)
    
    # Filter data for the current PTNR_RM_CODE and select only 'WEEK' and 'SALES' columns
    df_code = df[df['PTNR_RM_CODE'] == code][['WEEK', 'SALES']]
    
    # Splitting data into train and test
    train_size = int(0.9 * len(df_code))
    df_train = df_code[:train_size]
    df_test = df_code[train_size:]
    
    cols = df_code.columns

    


    best_order = 0
    best_model = None
    best_mape = float('inf')

    for i in range(1, max_lags + 1):
        print('Order =', i)
        model = VAR(df_train)
        results = model.fit(i)


        lag_order = results.k_ar
        forecast = results.forecast(df_train.values[-lag_order:], len(df_test))

        mape_val = mean_absolute_percentage_error(df_test[cols].values, forecast)
        print('MAPE:', mape_val)
        print('-' * 40)


   
        if mape_val < best_mape:
            best_mape = mape_val
            best_order = i
            best_model = results

    print('Best order:', best_order)

 
    print('Overall MAPE:', best_mape)
    print('\n')
```
