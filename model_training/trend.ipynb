```python
import numpy as np
```

```python
sales_matrix = np.load('/Users/harshitgupta/Desktop/vs /VS-Data-Den/data_cleaning/sales_matrix.npy')
```

```python
total_sales = np.sum(sales_matrix, axis=0)


dates = np.arange(len(total_sales))

```

```python
import matplotlib.pyplot as plt
```

```python
# Plot the trend
plt.figure(figsize=(10, 6))
plt.plot(dates, total_sales, color='blue')
plt.title('Total Sales Trend')
plt.xlabel('Time')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```
